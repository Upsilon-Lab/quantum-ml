```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("galaxy_distances.csv")
df = df.drop(0, axis=0)
df = df.replace("-", np.nan)
df = df.dropna()
df['distance'] = pd.to_numeric(df['distance'])
df['velocity'] = pd.to_numeric(df['velocity'])
```

```python
sns.scatterplot(data = df, x = 'distance', y = 'velocity')
```

```python
from sklearn import datasets, linear_model
distance = df['distance'].values.reshape(-1,1)
velocity = df['velocity'].values.reshape(-1,1)
regr = linear_model.LinearRegression()
regr.fit(distance, velocity)
y_pred = regr.predict(distance)
plt.plot(distance, y_pred)
plt.scatter(distance, velocity, c = 'red', s=3)
plt.xlabel('distance')
plt.ylabel('velocity')
print("The Hubble constant is", float(regr.coef_), "km/s/Mpc")
```
