```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def genPot(spikiness, length=100, seed=None):
    np.random.seed(seed)
    pot = np.zeros(length)
    count = 0
    gen = np.random.rand()

    while count < length:
        if np.random.rand() < spikiness:
            gen = np.random.rand()
        pot[count] = gen
        count += 1

    return pot

plt.plot(genPot(0.3))
print ("Yay! We have generated a random potential:")
            
```
