```
import numpy as np
import matplotlib.pyplot as plt

#Endpoint sets the x endpoint 
endpoint = 100
numpoints = 100
x = np.linspace(1,endpoint,numpoints)

amp = 5
period= 15

a = np.random.random(numpoints)
p = np.random.random(numpoints)

asingle = np.random.random(1)
psingle = np.random.random(1)


#f1 and f2 show the original triangle and sawtooth waves 
#f1rand and f2rand use the rand function to generate slightly different period and amplitudes for each point
#f1randsingle and f2randsingle generate a single random amplitude for period. This way the shape of the sawtooth and 
# triangle waves aren't changed. 

f1 = 2*amp / np.pi * np.arcsin(np.sin(2*np.pi *x/period))
f2 = -2*amp / np.pi * np.arctan(np.tan(np.pi *x/period))

f1rand = 2*a / np.pi * np.arcsin(np.sin(2*np.pi *x/p))
f2rand = -2*a / np.pi * np.arctan(np.tan(np.pi *x/p))

f1randsingle = 2*asingle / np.pi * np.arcsin(np.sin(2*np.pi *x/psingle))
f2randsingle = -2*asingle / np.pi * np.arctan(np.tan(np.pi *x/psingle))


plt.figure(1)
plt.plot(x, f1)

plt.figure(2)
plt.plot(x, f2)

plt.figure(3)
plt.plot(x, f1rand)

plt.figure(4)
plt.plot(x,f2rand)

plt.figure(5)
plt.plot(x, f1randsingle)

plt.figure(6)
plt.plot(x,f2randsingle)
```
